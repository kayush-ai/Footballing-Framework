```python
def read_csv(name):
    df = pd.read_csv(name)
    df.columns = df.columns.str.strip()
    for c in df.columns:
        if c in ['Rk', 'Matches', 'Born']:
            df = df.drop(columns=c)
        if c == 'Squad':
            df[c] = df[c].str.replace(r"^\S+\s+", '', regex=True)

    return df
```

#### Scaling

```python
from sklearn.preprocessing import MinMaxScaler

defence_scaled = defence.copy()
```

```python
positive_metrics = [
    'TklW', 'Int', 'Blocks', 'Tkl+Int', 'Clr',
    'Cmp%', 'Pass_PrgDist', 'PrgP', 'Pass_1_3',
    'PPA', 'PrgC', 'Carry_PrgDist', 'Succ%', 'SCA90',
    'GCA90', 'xA'
]

negative_metrics = ['Err', 'Mis', 'Dis']
```

```python
minMax = MinMaxScaler()
defence_scaled[positive_metrics] = minMax.fit_transform(
    defence_scaled[positive_metrics]
)
```

```python
defence_scaled[negative_metrics] = minMax.fit_transform(
    defence_scaled[negative_metrics]
)

defence_scaled[negative_metrics] = 1 - defence_scaled[negative_metrics]
```

```python
index_metrics = positive_metrics + negative_metrics
defence_scaled['Defence_Index'] = defence_scaled[index_metrics].mean(axis=1)
```

```python
defence_scaled['Defence_Index'] = minMax.fit_transform(
    defence_scaled[['Defence_Index']]
)
```

```python
defence_scaled.head()
```

***All metrics were min–max scaled and directionally adjusted to ensure fair comparison across defensive, progressive, and possession-based contributions before constructing a composite Defender Index***

```python
stopper_metrics = [
    'TklW', 'Int', 'Blocks', 'Clr', 'Tkl+Int',
    'Err', 'Mis', 'Dis'
]
defence_scaled['Stopper_score'] = defence_scaled[stopper_metrics].mean(axis=1)
```

    Interpretation:
    
    0.75+ → Elite stopper
    0.55–0.75 → Reliable defensive option
    <0.55 → Not stopper-oriented

```python
progressor_metrics = [
    'Cmp%', 'Pass_PrgDist', 'PrgP',
    'Pass_1_3', 'PPA',
    'PrgC', 'Carry_PrgDist', 'Succ%'
]
defence_scaled['Progressor_Score'] = defence_scaled[progressor_metrics].mean(axis=1)
```

    Interpretation:
    
    0.75+ → Elite build-up defender
    0.55–0.75 → Support progressor
    <0.55 → Limited on-ball impact

```python
hybrid_metrics = (
    stopper_metrics +
    progressor_metrics +
    ['SCA90', 'GCA90', 'xA']
)
defence_scaled['Hybrid_Score'] = defence_scaled[hybrid_metrics].mean(axis=1)
```

    Interpretation:

    0.75+ → Complete defender
    0.60–0.75 → Tactical flexibility asset
    <0.60 → Role-specific only

```python
defence_scaled.head()
```

```python
def assign_archetype(row):
    score = {
        'Stopper': row['Stopper_score'],
        'Progressor': row['Progressor_Score'],
        'Hybrid': row['Hybrid_Score']
    }
    return max(score, key=score.get)

defence_scaled['Defender_Archetype'] = defence_scaled.apply(assign_archetype, axis=1)
```

***Defenders were classified into Stopper, Progressor, and Hybrid archetypes using scaled performance metrics across defensive actions, ball progression, and chance creation to reflect modern tactical roles***

***Columns were selected based on their contribution to control, progression, chance creation, and risk management, rather than raw volume or descriptive event types***

```python
scale_cols = [
    # Defensive
    'Tkl', 'TklW', 'Tkl%', 'Int', 'Tkl+Int',
    # Creation
    'SCA90', 'GCA90', 'xA', 'KP', 'PPA',
    # Passing
    'Cmp%', 'Pass_PrgDist', 'PrgP', 'Pass_1_3',
    # Carrying
    'Carries', 'Carry_PrgDist', 'PrgC', 'Succ%',
    # Shooting efficiency
    'xG', 'npxG', 'npxG/Sh', 'G-xG',
    # Risk (Negative directional)
    'Err', 'Mis', 'Dis', 'Tkld', 'Tkld%'
]

mid_scaled = mid.copy()
minMaxMid = MinMaxScaler()
mid_scaled[scale_cols] = minMaxMid.fit_transform(mid[scale_cols])
```

```python
mid_scaled[scale_cols].describe().loc[['min','max']]
```

```python
negative_metrics = ['Err', 'Mis', 'Dis', 'Tkld', 'Tkld%']

for col in negative_metrics:
    mid_scaled[col] = 1 - mid_scaled[col]
```

***Negative-direction metrics were min–max scaled and then inverted to ensure all features followed a consistent ‘higher-is-better'***

```python
controller_cols = [
    'Tkl', 'TklW', 'Tkl%', 'Int', 'Tkl+Int',
    'Rec', 'Touches', 'Pass_Cmp', 'Cmp%',
    'Err', 'Mis', 'Dis'
]

mid_scaled['Controller_Score'] = mid_scaled[controller_cols].mean(axis=1)
```

    Interpretation
* High score = safe, reliable, press-resistant midfielder
* Low score = chaos merchant under pressure

```python
progressor_cols = [
    'PrgP', 'Pass_PrgDist', 'Pass_1_3', 'PPA',
    'Carries', 'Carry_PrgDist', 'PrgC',
    'Carry_1_3', 'CPA', 'PrgR'
]

mid_scaled['Progressor_Score'] = mid_scaled[progressor_cols].mean(axis=1)
```

    Interpretation
* High score = line-breaking machine
* Low score = sterile possession

```python
creator_cols = [
    'SCA90', 'GCA90', 'KP', 'xA', 'xAG',
    'TB', 'Sw', 'Crs', 'CrsPA',
    'Shots', 'SoT', 'xG', 'npxG',
    'npxG/Sh', 'G-xG', 'np:G-xG'
]

mid_scaled['Creator_Score'] = mid_scaled[creator_cols].mean(axis=1)
```

    Interpretation
* High score = final-third value generator
* Low score = tidy but toothless

```python
mid_scaled['Midfielder_Index'] = (
    mid_scaled['Controller_Score'] +
    mid_scaled['Progressor_Score'] +
    mid_scaled['Creator_Score']
) / 3
```

```python
scaler = MinMaxScaler()

mid_scaled[['Controller_Score_n',
    'Progressor_Score_n',
    'Creator_Score_n',
    'Midfielder_Index_n']] = scaler.fit_transform(
        mid_scaled[['Controller_Score', 'Progressor_Score', 'Creator_Score', 'Midfielder_Index']]
)
```

```python
def label_midfielder(row, threshold=0.7):
    scores = {
        'Controller': row['Controller_Score_n'],
        'Progressor': row['Progressor_Score_n'],
        'Creator': row['Creator_Score_n']
    }
    
    sorted_scores = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    
    top_role, top_value = sorted_scores[0]
    second_role, second_value = sorted_scores[1]
    
    if top_value >= threshold and second_value >= threshold:
        return 'Hybrid'
    else:
        return top_role
```

```python
mid_scaled['Midfielder_Archetype'] = mid_scaled.apply(label_midfielder, axis=1)
```

```python
forward_sc = forwards.copy()
forscaler = MinMaxScaler()
forward_sc[['Forward_Index_scaled', 'Finisher_Score_sc', 'Volume_Score_sc', 'Link_Score_sc']] = forscaler.fit_transform(
    forwards[['Forward_Index', 'Finisher_Score', 'Volume_Score', 'Link_Score']]
)
```

```python
forward_sc.head()
```

```python
def label_forward(row, threshold=0.7):
    scores = {
        'Finisher': row['Finisher_Score_sc'],
        'Volume Attacker': row['Volume_Score_sc'],
        'Link Forward': row['Link_Score_sc']
    }
    
    sorted_scores = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    
    top_role, top_value = sorted_scores[0]
    second_role, second_value = sorted_scores[1]
    
    if top_value >= threshold and second_value >= threshold:
        return 'Hybrid Forward'
    else:
        return top_role

forward_sc['Forward_Archetype'] = forward_sc.apply(label_forward, axis=1)
```
